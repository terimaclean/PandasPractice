```python
import pandas as pd
import matplotlib.pyplot as plt
poke = pd.read_csv("PokemonOneofEach.csv", index_col="Key")
```

#Sample of Pokemon Data
####Data was cleaned and imported into SQL where was a query was created to populate the csv used for this analysis.

#Top Ten Pokemon for Each Stat

###HP

```python
#Chart of top ten Pokemon by HP descending
topHP = poke.sort_values(['HP'], ascending=False).head(10)
plt.style.use("ggplot")
topHP.plot(y="HP", x="Name", kind="bar")
```

###Attack

```python
#Chart of top ten Pokemon by Attack descending
topAttack = poke.sort_values(['Attack'], ascending=False).head(10)
plt.style.use("ggplot")
topAttack.plot(y="Attack", x="Name", kind="bar")
```

###Defense

```python
#Chart of top ten Pokemon by Defense descending
topDefense = poke.sort_values(['Defense'], ascending=False).head(10)
plt.style.use("ggplot")
topDefense.plot(y="Defense", x="Name", kind="bar")
```

###Special Attack

```python
#Chart of top ten Pokemon by Special Attack descending
topSpatk = poke.sort_values(['Sp. Atk'], ascending=False).head(10)
plt.style.use("ggplot")
topSpatk.plot(y="Sp. Atk", x="Name", kind="bar")
```

###Special Defense

```python
#Chart of top ten Pokemon by Special Defense descending
topSpdef = poke.sort_values(['Sp. Def'], ascending=False).head(10)
plt.style.use("ggplot")
topSpdef.plot(y="Sp. Def", x="Name", kind="bar")
```

###Total

```python
#Chart of top ten Pokemon by total stats descending
topTotal = poke.sort_values(['Total'], ascending=False).head(10)
plt.style.use("ggplot")
topTotal.plot(y="Total", x="Name", kind="bar")
```
